# Training models on Floydhub

What is Floydhub?

- Easy way to train your Deep Learning models on cloud GPUs if you don't have one already
- Cheap cloud GPUs
- Trial GPU hours


How do you get started?

1) Install Floydhub CLI:  https://docs.floydhub.com/guides/basics/install/

2) Understand core concepts
  - Project: this is where your code is will live
  - Jobs: Actual GPU/CPU jobs, charged per usage
  - Datasets: Your dataset (e.g. catvsdog images). You must upload them using the Floydhub CLI
  - Environments: GPU and CPU environments are available

# Cheatsheet

Upload a dataset and start

environment

1) Bootstrapping

```bash
# Authenticate your current floydhub cli
floyd login

# Initiate floydhub local project
floyd init

```

2) Bootstrapping floydhub remote project and datasets

- Upload dataset by following: https://docs.floydhub.com/guides/create_and_upload_dataset/
  
  In short, once the remote datasets is created
  ```bash
  # 1. cd into the dataset directory
  
  # 2. Initiate the dataset
  floyd data init <datasetname>
  
  # 3. Upload
  floyd data upload
  ```
  

3) Starting a Jupyter notebook with gpu mode

- make sure you already have .floydignore file

```bash
# cd into "deepcollege" root directory

floyd run --gpu --mode jupyter --data visualbbasic/humpback-whales:/inputs/
```


Please follow https://docs.floydhub.com/ for the full length tutorial on how to do this

4) You can also run an individual python file

By initiating the project, your Python file should already be uploaded to the project

Example command format:

```
floyd run \
    [--gpu] \
    --env <env> \
    --data <your_dataset>:<mounting_point_dataset> \
    --data <output_of_previous_job>:<mounting_point_model> \
    "python <script_and_parameters>"
```

Example command:

- GPU env
- TF 1.3
- Datasets

```
floyd run \
    --gpu \
    --env tensorflow-1.3 \
    --data redeipirati/datasets/mnist/1:input \
    --data <your-username>/projects/save-and-resume/<jobs>/output:/model \
    'python tf_mnist_cnn.py'
```
